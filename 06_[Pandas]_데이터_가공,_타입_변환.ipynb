# 타입 변환

```python
df['PassengerId'].head() # 연산을 하기 위한 숫자가 아닌 구분을 하기 위한 구분 값 (이때는 ㅇㄹ.["ㅖㅁㄴㄴ둫ㄷ걍정수를 사용) 두 종류를 잘 알고 데이터 분석을 해야 함
```

```python
df['PassengerId'].astype(str) # 원하는 데이터를 타입 캐스팅을 하는 것
```

```python
# 실제 데이터에 타입 변환 적용
df['PassengerId'] = df['PassengerId'].astype(str) # object는 모든 클래스의 최상위 클래스 모든 데이터를 처리하기 위해서 사용함 그래서 object를 사용해도 되지만 명확하게 하기 위해서 str 사용
df.info()
```

```python
# 어떠한 데이터(숫자)에 nan값이 들어있으면 무조건 실수 취급함 ex) Age
df['Age'].head()
```

```python
# 시리즈에 NaN이 포함되면 타입변경 불가능!
df['Age'].astype(int).head() # 에러남 nan값이 있으면 타입 변경이 안됨 반드시 nan값을 처리 해줘야함
```

```python
# NaN 처리하기
df["Age"].fillna(-1).astype(int).head()
# 나이에 0과 -1을 넣는 것은 차이가 있음 0살은 의미가 있음 -1은 음수인 나이가 있을 수 없기에 의미가 없어서 임시의 값이라는 것을 알 수 있음
```

```python
df['Age'] = df['Age'].fillna(-1).astype(int)
df.tail() # 실수에서 정수로 바뀐 것을 볼 수 있고 NaN값이 -1로 바뀐 것도 볼 수 있음
```

```python
# numpy를 이용해 nan 지정 가능
import numpy as np

df['Age'] = df['Age'].fillna(-1).astype(int).replace(-1, np.nan) # 치환의 개념 replace
df.tail()
```
