```python
import pandas as pd
```

```python
df = pd.read_csv("./data/titanic_train.csv")
df.head()
```

```python
# info()에서 확인
df.info() # 전체 데이터가 891개인데 non-null값이 891이면 결측치가 없는 것이다.
```

```python
# 결측값을 찾기위해 isnull() 함수를 사용해도 무관!(옛날 버전)
#   - 예전에는 문자열의 결측치는 null로 표기가 됐던 시절이 있었음..ㅠㅠ
```

# 결측값 제거
- `dropna`

```python
# NaN값이 1개라도 들어있는 행을 삭제
df.dropna() # 기본 axis=0
# nan값이 한개라도 포함이 되어있으면 다 지워지는 것
```

```python
# NaN값이 1개라도 들어있는 컬럼을 삭제
df.dropna(axis=1)
```

```python
# how 옵션을 이용하면 NaN의 포함 여부를 설정 할 수 있다.
#  기본 : any -> 1개라도 들어있으면 다 지우기
#        all -> 모든 데이터가 NaN인 경우 지우기

df.dropna(how='all') # 모든 컬럼의 데이터가 NaN인 경우 삭제
```

```python
# 선택한 컬럼 중 NaN이 포함되어 있으면 지우게 하기
df.dropna(subset=["Cabin", "Age"]) # Cabin과 Age에 하나라도 NaN이 있으면 행 삭제
```

# 결측값 채우기(대치)

```python
df.tail()
```

```python
# NaN값을 일괄적으로 채우기
df.fillna(-1).tail()
```

```python
# 나이의 평균으로 NaN값 채우기
age_mean = df['Age'].mean()
age_mean
```

```python
df["age_fillna"] = df['Age'].fillna(age_mean)
df
```

```python
# nan값 위나 아래로 채우는 경우도 있음, 주가변동같은 경우(시계열 데이터!) 평균보다는 그게 더 낫기 때문에
df.fillna(method='ffill').tail() # 위의 데이터로 채우는 경우
```

```python
df.fillna(method='bfill').tail() # 아래의 데이터로 채우는 경우
```
