Series는 한 항목의 Data들 (배열 생각) target이 여러개 있는 거

DataFrame은 그런 Series들을 모아놓은 것 (2차원 배열 생각)

```python
import pandas as pd
```

# 시리즈 생성하기

```python
# dict를 이용해 Series 만들기
sample_dict = {"a": 1, "b": 2, "c": 3}
dict_series = pd.Series(sample_dict)
dict_series
# a,b,c.. 는 row index, 위에 0,1,2...은 column index
```

```python
lst = [1, 2, 3]
list_series = pd.Series(lst, index=["a", "b", "c"])
list_series
```

```python
# 인덱스를 따로 지정하지 않으면 자동으로 range index(마치 offset index 같은 것)
list_series2 = pd.Series(lst)
list_series2
```

```python
list_series3 = pd.Series(lst, name="컬럼의 이름!")
list_series3
```

```python
# 컬럼과 시리즈는 거의 같은 거라고 생각해도 됨
```

# 데이터프레임
- 시리즈를 모아 놓은 것

```python
sample_dict = {
    'Name': ["A", "B", "C"],
    'height': [180.5, 173.1, 178.3], # 모든 데이터의 개수가 같아야 만들어짐 아니면 오류
}

df = pd.DataFrame(sample_dict)
df
```

```python
sample_dict = {
    'Name': ["A", "B", "C", "D"],
    'height': [180.5, 173.1, 178.3], # 모든 데이터의 개수가 같아야 만들어짐 아니면 오류
}

df = pd.DataFrame(sample_dict)
df
```

```python
lst = [['A', 180.3],
       ['B', 173.6],
       ['C', 178.3]] # 리스트를 만들 때 2차원 배열 모양으로 만들어야 데이터 프레임이 만들어 짐

df = pd.DataFrame(lst, columns=["Name", "Height"])
df
```

```python
lst = [['A', 180.3],
       ['B', 173.6],
       ['C', 178.3, 4]] # 이러면 오류, 항목의 데이터 개수는 항상 동일해야 함

df = pd.DataFrame(lst, columns=["Name", "Height"])
df
```

```python
# 시리즈를 합쳐서 데이터프레임 만들기
name_series = pd.Series(["A", "B", "C"])
height_series = pd.Series([180.3, 175.3, 178.3])

df = pd.DataFrame({"name": name_series, 'height': height_series})
```

```python
df
```

```python
# JSONArray 형식으로 된 데이터를 데이터프레임으로 만들기. (진짜 많이 사용합니다.)
json_sample_array = [
    {"name" : "A", 'height': 180.3},
    {"name" : "B", 'height': 178.3},
    {"name" : "C", 'height': 175.3}
]

df = pd.DataFrame(json_sample_array)
df
```
