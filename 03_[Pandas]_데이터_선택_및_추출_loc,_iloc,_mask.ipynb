## loc, iloc
- `loc` : 이름 값을 활용해 조회

```python
df.loc[3] # 행과 열의 접근을 도움 !3번! 행의 데이터가 나오게 됨 3번째가 아님!
# df안에서는 ~번 index로 읽어야함 번째 안됨
```

```python
# loc[행 인덱스, 열 인덱스]
df.loc[3, 'Survived']
```

```python
df.loc[3:5, 'Name'] # loc은 offset index가 아니기 때문에 3번 ~ 5번의 탑승객 정보가 나오게 됨
```

```python
# 3 ~ 10 번 탑승객의 Name, Fare, Pclass 조회
columns = ["Name", "Fare", "Pclass"] # fancy indexing을 사용해야함
df.loc[3:10, columns]
```

```python
# 1, 3, 5번 탑승객의 Age, Pclass
rows = [1, 3, 5]
columns = ["Age", "Pclass"]

df.loc[rows, columns]

# columns = ["Age", "Pclass"] # fancy indexing을 사용해야함
# df.loc[1:5:2, columns]
```

## loc를 이용한 조건 추출
- 행(레코드)에 대한 조건을 설정하여 원하는 조건에 맞는 행만 추출

```python
# 조건 마스크
mask = df['Pclass'] == 1
mask # 1인지 아닌지만 검사한 목록만 나오게 됨
```

```python
# 조건 마스크를 행에다 씌워서 True인 것만 추출
df.loc[mask] # True인 행만 나오게 됨
```

```python
# AND, OR 조건
mask = (df["Pclass"] == 1) | (df["Age"] >= 30) # & : AND 조건, | : OR 조건
mask
```

```python
df.loc[mask] # pclass가 1거나 age가 30이상인 사람들만 나오게 됨
```

```python
# query : 문자열을 이용해 조건절을 지정
df.query('Pclass == 1')
```

```python
df.query('Pclass == 1 and Age >= 30')
```

```python
# query에서 열 지정하기
df.query('Pclass == 1 and Age >= 30')[["Name", "Age", "Pclass"]] # 조회를 한 상태에서 열 선택
```

```python
# isin 조건
mask = df["PassengerId"].isin([3, 100, 500])
df.loc[mask] # 이렇게만 해도 나오긴 하지만 df.loc[mask]처럼 웬만하면 loc 쓰는게 더 좋음
```

```python
df.query("PassengerId in [3,100,500]")
```

```python
df.query("PassengerId.isin([3,100,500])")
```

```python
# 변수에 조회할 조건을 넣어서 query로 조회
passenger_samples = [3, 100, 500]

df.query("PassengerId in @passenger_samples") # 조건을 넣어서 조회는 @를 넣어서 하기
```
