## **문자형을 날짜형으로 변경**
- 날짜가 문자형으로 되어있다면 날짜형으로 변경해야 여러가지 날짜 계산을 할 수 있습니다.
- `pd.to_datetime(컬럼, format='날짜 형식')`

|형식|설명|
|:--|:--|
|%Y|0을 채운 4자리 연도|
|%y|0을 채운 2자리 연도|
|%m|0을 채운 월|
|%d|0을 채운 일|
|%H|0을 채운 시간|
|%M|0을 채운 분|
|%S|0을 채운 초|

```python
df['Date1'] = pd.to_datetime(df['Date'], format='%Y-%m-%d') # dtype: datetime64로 바뀐 것을 알 수 있음
df.info()
```

# 날짜 데이터를 원하는 형식으로 변경

```python
df['Date1'].head()
```

```python
df['Date1'].dt.strftime("%Y-%m") # 원하는 형식으로 바꿔줌
```

```python
df['Date1'].dt.strftime("%Y년 %m월")
```

```python
df['year'] = df['Date1'].dt.year # 년도
df['month'] = df['Date1'].dt.month # 월
df['day'] = df['Date1'].dt.day # 일

df['dayofweek'] = df['Date1'].dt.dayofweek # 요일( 0 ~ 6 )
df['dayname'] = df['Date1'].dt.day_name()

df.head()
```

# 날짜 계산
- day 연산: `pd.Timedelta(day=숫자)`
- month 연산: `DateOffset(months=숫자)`
- year 연산: `DateOffset(years=숫자)`

```python
# 하루 뒤 날짜 계산하기
df['plus day1'] = df['Date1'] + pd.Timedelta(days=1)
df.head()
```

```python
# 7일 뒤 날짜 계산하기
df['plus day7'] = df['Date1'] + pd.Timedelta(days=7)
df.head()
```

```python
# 7일 전 날짜 계산하기
df['minus day7'] = df['Date1'] - pd.Timedelta(days=7)
df.head()
```

```python
# DateOffset 사용하기 - 년, 월에 대한 변화
from pandas.tseries.offsets import DateOffset

df['plus month1'] = df['Date1'] + DateOffset(months=1)
df.head()
```

```python
df['minus month3'] = df['Date1'] - DateOffset(months=3)
df.head()
```

```python
df['plus year1'] = df['Date1'] + DateOffset(years=1)
df.head()
```

# 기간 이동 계산
- rolling

지정한 기간 만큼 시간을 하나의 컨텍스트로 보는 것

```python
df = pd.read_csv("./data/temperatures.csv")
df.head()
```

```python
# rolling : 지정한 기간 만큼 시간을 하나의 컨텍스트로 보는 것
# 컨텍스트는 맥락, 지정한 범위 만큼을 하나로 묶는 것
df["ma7"] = df["Temp"].rolling(7).mean()
# 7개의 데이터가 한개가 됨 (집계) 여러개의 값을 하나의 값으롬 만들어 주는 작업 ex) 평균, 중앙값, 표준편차... 여러개의 숫자를 하나의 데이터로 대표할 수 있는 작업
df.head(20) # 이동평균
# 앞의 7개만 묶어지는 박스 (한칸씩 이동한)를 컨텍스트 윈도우라고 함
```

```python
df['ma30'] = df['Temp'].rolling(30).mean()
df.head(30) # rolling 값을 너무 크게 잡으면 변화가 없을 수 도 있고 작게 잡으면 트렌드를 알 기 힘들 수 있음 적당한 윈도우를 잡는 것이 중요
```

# 행 이동

```python
df['Time Shift1'] = df['Temp'].shift(1) # 각 행의 Temp 데이터를 한칸씩 아래로 이동
df.head()
```

```python
# 오늘 날짜와 어제 날짜의 온도 변화 % 구하기
df['pct change'] = (df['Time Shift1'] - df['Temp']) / df['Temp']
df.head()
```
