# Concat

## row 기준 합치기
- 행 인덱스를 기준으로 합치기

```python
df1_copy = df1.copy()
```

```python
df1_concat = pd.concat([df1, df1_copy], axis=0) # axis=0이니까 아래쪽으로 데이터가 늘어 남 위에서 아래로 합쳐지는 것이기 때문에 columns의 이름이 같아야함
df1_concat
```

```python
# 인덱스 재 설정
df1_concat.reset_index(drop=True) # drop=True를 사용하면 인덱스가 재 세팅 됨
```

# merge
- `concat`과 `merge`는 단순하게 합쳐지는 목적 또는 특정 기준(`index`)으로 합치느냐에 따라서 용도가 달라집니다.
- `concat` : `row`나 `column` 기준으로 단순하게 이어 붙이기
- `merge` : 특정한 고유 키(`unique id`) 값을 기준으로 **병합**

**`pd.merge(left, right, on='기준 컬럼', how='left')`**
- `left`, `right` : 병합할 두 DataFrame
- `on` : 병합의 기준이 되는 컬럼의 이름
- `how` : `left, right, inner, outer` 4가지 병합방식 설정

```python
df2_right = df2.drop([1, 3, 5, 7, 9], axis=0)
df2_right = df2_right.reset_index(drop=True)
df2_right
```

```python
df1.head()
```

```python
pd.concat([df1, df2_right], axis=1) # drop된 것 때문에 합쳐진 값이 이상한 것을 알 수 있다
```

```python
pd.merge(
    df1,
    df2_right,
    on='이름',
    how='left' # left에 위치한 df1이 기준이 된다. 기준이 된 데이터 프레임은 데이터의 삭제가 일어나지 않는다.
)
```

```python
pd.merge(
    df1,
    df2_right,
    on='이름',
    how='right' # right에 위치한 df1이 기준이 된다. 오른쪽 데이터가 기준이 되기 때문에 오른쪽 데이터에 맞춰져서 왼쪽에디터에 변경이 일어났다.
)
```

# inner, outer 방식
- `inner` : 두 데이터 프레임에 **모두** 키값이 존재하는 경우 병합
- `outer` : 하나의 데이터 프레임에 키값이 존재하는 경우 모두 병합
  - 없는 값은 NaN값으로 대입

교집합, 합집합

```python
df1_left = df1.drop([1, 2, 6, 8])
df1_left
```

```python
pd.merge(df1_left, df2_right, on="이름", how='inner')
# 겹치는 것만 살아남게 됨 교집합(inner)
```

```python
pd.merge(df1_left, df2_right, on="이름", how='outer')
# 양쪽의 데이터가 모조리 살아남게 됨 합집합 (outer) 없는 부분은 NaN값으로 채워짐
```
